## Постановка задачи и описание данных

Для решения бизнес задачи маркетплейсу необходимо определять похожие товары. Для решения этой задачи предлагается следующее:
* разработать алгоритм, который для всех товаров из размеченных экспертами предложит несколько вариантов наиболее похожих товаров из всех товаров маркетплейса.
* оценить качество поиска схожих товаров с помощью метрики _accuracy@5_.

Исходные данные:
* base.csv - анонимизированный набор товаров. Каждый товар представлен как уникальный id (0-base, 1-base, 2-base) и вектор признаков размерностью 72.
* train.csv - обучающий датасет. Каждая строчка - один товар, для которого известен уникальный id (0-query, 1-query, …) , вектор признаков И id товара из base.csv, который максимально похож на него (по мнению экспертов).
* validation.csv - датасет с товарами (уникальный id и вектор признаков), для которых надо найти наиболее близкие товары из base.csv
* validation_answer.csv - правильные ответы к предыдущему файлу.

План работы:
1. Загрузка данных и разведочный анализ.
2. Обработка признаков.
3. Подбор параметров модели с целью увеличения метрики.
4. Финальный расчет на валидационных данных.
5. Вывод.

## Импорт необходимых библиотек

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.cluster import KMeans

import faiss

from time import time
from tqdm import tqdm
```

## Первый взгляд на данные

```python
df_base = pd.read_csv(r'base.csv', index_col=0)
df_train = pd.read_csv(r'train.csv', index_col=0)
df_valid = pd.read_csv(r'validation.csv', index_col=0)
df_valid_ans = pd.read_csv(r'validation_answer.csv', index_col=0)
```

Функция, выводящая обшую информацию о датасете:

```python
# выделим часть данных для изучения
NUM_OF_SAMPLES = 2 * 10 **4

def first_look(df):
    
    display(df.head())
    
    display(df.info())
    
    df.sample(NUM_OF_SAMPLES).hist(figsize=(20, 20), bins=50)
```

```python
first_look(df_base)
```

```python
first_look(df_train)
```

```python
first_look(df_valid)
```

Пропусков нет - уже неплохо.

Выделим подозрительные признаки:

```python
suspect_features = ['6', '21', '25', '33', '44', '59', '65', '70']
```

```python
df_valid_ans.head()
```

Выделим признаки и таргет:

```python
targets = df_train["Target"]
df_train.drop("Target", axis=1, inplace=True)
```

## Бейзлайн с Faiss

```python
dims = df_base.shape[1]
n_cells = 1 
quantizer = faiss.IndexFlatL2(dims)
idx_l2 = faiss.IndexIVFFlat(quantizer, dims, n_cells)
```

```python
idx_l2.train(np.ascontiguousarray(df_base.values[:NUM_OF_SAMPLES, :]).astype('float32'))

idx_l2.add(np.ascontiguousarray(df_base.values).astype('float32'))
```

```python
base_index = {k: v for k, v in enumerate(df_base.index.to_list())}
```

```python
r, idx = idx_l2.search(np.ascontiguousarray(df_train.values).astype('float32'), 5)
```

```python
acc = 0
for target, el in zip(targets.values.tolist(), idx.tolist()):
    acc += int(target in [base_index[r] for r in el])
```

```python
print(f"Метрика accuracy@5 на базовой модели = {100 * acc / len(idx)}")
```

## Обработка признаков для улучшения метрики accuracy@5

Проведем масштабирование признаков:

```python
sclr = StandardScaler()

df_base_scaled = pd.DataFrame(data=sclr.fit_transform(df_base),
                             index=df_base.index,
                             columns=df_base.columns)

df_train_scaled = pd.DataFrame(data=sclr.transform(df_train),
                             index=df_train.index,
                             columns=df_train.columns)

df_valid_scaled = pd.DataFrame(data=sclr.transform(df_valid),
                             index=df_valid.index,
                             columns=df_valid.columns)
```

Методом локтя подберем оптимальное число класстеров:

```python
inertias = []
K = range(10, 301, 20)

for k in K:
    kmean_model = KMeans(n_clusters=k, max_iter=1000)
    kmean_model.fit(df_train_scaled.sample(NUM_OF_SAMPLES))
    
    inertias.append(kmean_model.inertia_)
    
plt.plot(K, inertias, 'o-')
plt.xlabel('Количество кластеров')
plt.ylabel('SSE')
plt.title('Метод локтя')
plt.grid()
plt.show()
```

Оптимальное число кластеров примем равным 250.

Это должно сильно сократить время вычислений, так что попробуем взять весь _base_.

```python
dims = df_base_scaled.shape[1]
n_cells = 250 
quantizer = faiss.IndexFlatL2(dims)
idx_l2 = faiss.IndexIVFFlat(quantizer, dims, n_cells)
```

```python
%%time

idx_l2.train(np.ascontiguousarray(df_base_scaled.values).astype('float32'))

idx_l2.add(np.ascontiguousarray(df_base_scaled.values).astype('float32'))
```

```python
base_index = {k: v for k, v in enumerate(df_base_scaled.index.to_list())}
```

```python
%%time

r, idx = idx_l2.search(np.ascontiguousarray(df_train_scaled.values).astype('float32'), 5)
```

```python
%%time

acc = 0
for target, el in zip(targets.values.tolist(), idx.tolist()):
    acc += int(target in [base_index[r] for r in el])
```

```python
print(f"Метрика accuracy@5 модели c отмасштабированными признаками и разбиением на 250 кластеров = {100 * acc / len(idx)}")
```

На масштабированных данных метрика значительно возросла!

Попробуем удалить некоторые признаки. Еще раз посмотрим на подозрительные:

Проверка:

```python
dims = df_base_scaled_dr.shape[1]
n_cells = 250 
quantizer = faiss.IndexFlatL2(dims)
idx_l2 = faiss.IndexIVFFlat(quantizer, dims, n_cells)
```

```python
%%time

idx_l2.train(np.ascontiguousarray(df_base_scaled_dr.values).astype('float32'))

idx_l2.add(np.ascontiguousarray(df_base_scaled_dr.values).astype('float32'))
```

```python
base_index = {k: v for k, v in enumerate(df_base_scaled_dr.index.to_list())}
```

```python
%%time

r, idx = idx_l2.search(np.ascontiguousarray(df_train_scaled_dr.values).astype('float32'), 5)
```

```python
%%time

acc = 0
for target, el in zip(targets.values.tolist(), idx.tolist()):
    acc += int(target in [base_index[r] for r in el])
```

```python
print(f"Метрика accuracy@5 модели c масштабированием и удалением признаков = {100 * acc / len(idx)}")
```

Значительное улучшение метрики!

Для дальнейшего улучшения метрики реализуем поиск не только в границах одного кластера, к центройду которого товар наиболее близок, но и в нескольких соседних к нему кластерах.

```python
dims = df_base_scaled_dr.shape[1]
n_cells = 250 
n_probe = 10 # количество соседних кластеров
quantizer = faiss.IndexFlatL2(dims)
idx_l2 = faiss.IndexIVFFlat(quantizer, dims, n_cells)
idx_l2.nprobe = n_probe
```

```python
%%time

idx_l2.train(np.ascontiguousarray(df_base_scaled_dr.values).astype('float32'))

idx_l2.add(np.ascontiguousarray(df_base_scaled_dr.values).astype('float32'))
```

```python
base_index = {k: v for k, v in enumerate(df_base_scaled_dr.index.to_list())}
```

```python
%%time

r, idx = idx_l2.search(np.ascontiguousarray(df_train_scaled_dr.values).astype('float32'), 5)
```

```python
%%time

acc = 0
for target, el in zip(targets.values.tolist(), idx.tolist()):
    acc += int(target in [base_index[r] for r in el])
```

```python
print(f"Метрика accuracy@5 модели c масштабированием и удалением признаков и поиском по соседям = {100 * acc / len(idx)}")
```

Хороший результат. Протестируем эту модель на валидационных данных.
