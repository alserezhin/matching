Гистограммы признаков:

```python
df_samples = df_train[:10000]

df_samples.hist(figsize=[20, 20], bins=50);
```

```python
target_samples = df_samples['Target']
features_samples = df_samples.drop('Target', axis=1)
```

## Pipeline KMeans + KNN

```python
sclr_std = StandardScaler()

scaled_features_samples = pd.DataFrame(data = sclr_std.fit_transform(features_samples),
                          index = features_samples.index,
                          columns = features_samples.columns)
```

```python
scaled_features_samples.head()
```

Определим оптимальное число кластеров:

```python
inertias = []
K = range(10, 300, 10)

for k in K:
    kmean_model = KMeans(n_clusters=k, max_iter=1000)
    kmean_model.fit(scaled_features_samples)
    
    inertias.append(kmean_model.inertia_)
    
plt.plot(K, inertias, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.show()
```

Примем оптимальное число кластеров равное 200.

```python
kmean_model = KMeans(n_clusters=200, max_iter=1000)
kmean_model.fit(features_samples)
```
